# Docstring parameter description Analysis
The length of the paramerer description are growing slowly in length. The code below will investigate if there is a problem

```python
df_param = pd.read_csv("../output/ml_inputs/_ml_param.csv")
```

```python
df_param
```

## Parameter comment length

```python
df_param['arg_comment_length'] = df_param['arg_comment'].dropna().apply(lambda x: x.split()).apply(len)
total = len(df_param['arg_comment'].dropna())
column = df_param['arg_comment_length']
column.describe()
```

### Above limit of 50

```python
df_limit = df_param.loc[df_param['arg_comment_length'] > 50]
```

#### Filter possible code clones

```python
df_limit = df_limit.drop_duplicates(subset=['arg_name', 'arg_comment'])
```

```python
df_limit
```

```python
df_limit.shape[0]
```

```python
limit_indexes = list(df_limit.index.values)
```

##NOTE
For finding repository info, a change had been made to the generate_df to include this info. However, this is not pushed
because it is not needed in the pipeline normally.

```python
df_param_repo_info = pd.read_csv("../output/ml_inputs/_ml_param_repo_info.csv")
```

```python
df_param_repo_info_filtered = df_param_repo_info.iloc[limit_indexes]
```

```python
df_param_repo_info_filtered
```

```python
df_param_repo_info_filtered['file'].tolist()
```

Based on taking random repo's from the list above, the conclusion is that long parameter comments either include examples or they explicitly describes the conditions on a parameter (e.g. a function parameter should have certain fields).

To remedy this, the only possible action is to take only the first sentence of a description since this should contain the overview information of that parameter.
